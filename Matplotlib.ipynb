```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
```

```python
x = np.linspace(0,100,100)
```

```python
plt.plot(x, 2*x, label='linear')
plt.plot(x, x**2, label='quadatric')
#plt.plot(x, x**3, label='cubic')

plt.xlabel('X')
plt.ylabel('Y')

plt.title('Polynomial Plot')

plt.legend()
plt.show()
```
